```python
import geopandas as gpd
import folium

# Cargar el archivo SHP en un GeoDataFrame
df_lol = gpd.read_file('E://shp//794551072264_s lolotla//130340001//conjunto_de_datos//130340001l.shp')

# Verificar que el GeoDataFrame se ha cargado correctamente
print(df_lol.head())

# Obtener el centro del GeoDataFrame para centrar el mapa
centro = df_lol.geometry.centroid.iloc[0].coords[0][::-1]

# Crear el mapa base centrado en la ubicación de interés
m = folium.Map(location=centro, zoom_start=12)

# Agregar el GeoDataFrame al mapa
folium.GeoJson(df_lol).add_to(m)

# Agregar capas de control
folium.LayerControl().add_to(m)

# Guardar el mapa en un archivo HTML
m.save('E://shp//mapa_lolotla.html')

# Mostrar el mapa
m
```
